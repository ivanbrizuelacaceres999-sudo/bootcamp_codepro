```python
#Escribir un programa en el que se pregunte al usuario por una temperatura (en grados Celsius), imprima la temperatura escrita, y luego clasificar la temperatura ingresada en los siguientes niveles:

#Entre -10 y 13 grados, que imprima "Ro'y koko".
#Entre 14 y 23 grados, que imprima "Refresca'i".
#Entre 24 y 28 grados, que imprima "Oikoite".
#Entre 29 y 45 grados, que imprima "Hakurooooo".
#Para todo lo demás, imprimir "Ñamano mba'e!".


def temperatura():
    
    temperatura_ingresada= int(input('ingrese la temperatura '))
    print('temperatura ingresa es :',temperatura_ingresada)
    
    if temperatura_ingresada >= -10 and temperatura_ingresada <= 13:
        print(' Ro  koko')
    if temperatura_ingresada >= 14 and temperatura_ingresada <= 23:
        print(" Refresca'i")
    if temperatura_ingresada >= 24 and temperatura_ingresada <= 28:
        print('Oikoitee')
    if temperatura_ingresada >= 29 and temperatura_ingresada <= 45:
        print(' Hakurooo')
    if temperatura_ingresada < -10 or temperatura_ingresada > 45:
        print(" Ñamano mba'e!")


temperatura()
```
